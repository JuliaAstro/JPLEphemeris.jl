# JPLEphemeris.jl Demonstration

```julia
using PyPlot
using Distance
using JPLEphemeris
```

```julia
# Load ephemeris DE421
eph = Ephemeris()

# 1st of January 2014
startepoch = 2456658.5
# Plus one Mars year
finalepoch = startepoch + 687.0
t = startepoch:finalepoch

# Positions of Earth-Moon-System's barycenter
earth = position(eph, "earthmoon", t)

# Positions of Mars
mars = position(eph, "mars", t);
```

```julia
d = colwise(Euclidean(),earth, mars)
min, minind = findmin(d)
max, maxind = findmax(d)
println("Closest approach of $min km at $(t[minind]).")
println("Maximum distance of $max km at $(t[maxind]).")
```

```julia
plot(vec(earth[1,:]), vec(earth[2,:]), "b")
plot(vec(mars[1,:]), vec(mars[2,:]), "r")
# Note: The Sun is not drawn to scale.
plot(0,0,"oy"; markersize=10)
plot([mars[1,minind], earth[1,minind]],[mars[2,minind], earth[2,minind]],"xc"; mew=3)
plot([mars[1,maxind], earth[1,maxind]],[mars[2,maxind], earth[2,maxind]],"xm"; mew=3)
xlim([-3.5e8,3e8])
ylim([-3e8,3e8])
xlabel("[km]")
ylabel("[km]")
annotate("2014-04-05", [mars[1,minind], mars[2,minind]]; textcoords="offset points", xytext=[-70, -20], color="c")
annotate("2015-07-11", [mars[1,maxind], mars[2,maxind]]; textcoords="offset points", xytext=[-40, 10], color="m")
```
